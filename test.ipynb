if-elif

If the user have to check for multiple condition and those condition determines what needs to be done. if-elif  is used.
Syntax: 

if condition1:
    #code to be executed if condition1 is true
elif condition2:
    #code to be executed if condition2 is true
elif condition3:
    #code to be executed if condition3 is true
else:
    #code to be executed if all the above condition doesnt matches

```python
marks = 55
if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
elif marks >= 60:
    print("Grade: D")
elif marks >= 50:
    print("Grade: E")
else:
    print("Grade: Fail")
print("End of program")
```

# nested if else
When there is a block of if-else, if, if-elif statement inside any if, if-else, if-elif then it is nested.

Syntax:
if condition1:
    #code to be executed if condition1 is true
elif condition:
    if condition2:
        #code to be executed if condition2 is true
    else:
        #code to be executed  if condition2 is false
else:
    #code to be executed if condition is false

```python
n = int(input("Enter number of subjects: "))

total = 0
fail_flag = False
distinction_count = 0

# Loop for marks input with validation
for i in range(1, n + 1):
    while True:
        marks = int(input(f"Enter marks for subject {i}: "))
        if 0 <= marks <= 100:
            break
        else:
            print("Invalid marks! Enter again (0–100).")

    total += marks

    if marks < 30:
        fail_flag = True

    if marks >= 80:
        distinction_count += 1

attendance = float(input("Enter attendance percentage: "))
category = input("Enter student category (regular/working): ").lower()

# Immediate failure condition
if attendance < 60 or fail_flag:
    result = "Fail"
    grade = "Fail"
    average = total / n

else:
    average = total / n

    # Grace marks rule
    if category == "regular" and 38 <= average < 40:
        average += 2

    # Grade calculation
    if average >= 85:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 55:
        grade = "C"
    elif average >= 40:
        grade = "D"
    else:
        grade = "Fail"

    # result = "Pass" if grade != "Fail" else "Fail"
    if grade != "Fail":
        result = "Pass"
    else:
        result = "Fail"

# Bonus distinction
if distinction_count >= 3 and result == "Pass":
    result += " with Distinction"

# Final Output
print("\n----- Final Result -----")
print("Total Marks:", total)
print("Average Marks:", round(average, 2))
print("Grade:", grade)
print("Result:", result)
```
